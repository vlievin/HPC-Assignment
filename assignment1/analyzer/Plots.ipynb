## Runtimes

```python
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline 
```

```python
colours = []
colours.append( (26, 188, 156))
colours.append( (52, 152, 219))
colours.append( (155, 89, 182))
colours.append( (241, 196, 15))
colours.append( (231, 76, 60))
colours.append( (46, 204, 113))
colours.append( (230, 126, 34))
colours.append( (149, 165, 166))
colours.append( (52, 73, 94))
```

```python
color=[]
for k in xrange(6):
        col = [ float(c) / 255.0 for c in colours[k]]
        color.append(col)
```

Matrix sizes (only the square case), for instance, m=n=k=500.

```python
size = [500, 1000, 1500, 2000]
```

Runtimes in seconds

```python
mkn = [4.050, 16.880, 57.250, 132.990]
mnk = [4.080, 21.840, 98.160, 154.730]
kmn = [4.110, 17.070, 57.820, 133.580]
knm = [3.410, 27.610, 115.520, 201.300]
nmk = [4.080, 17.370, 58.710, 137.330]
nkm = [5.380, 28.650, 115.640, 200.460]
```

```python
plt.rcParams['figure.figsize'] = (14.0, 8.0)
plt.title('Matrix size vs Runtime (sec)')
plt.xlabel('Matrix size ')
plt.ylabel('Runtime (sec)')
p1, = plt.plot( size, mkn, color = 'black', linewidth = 5, label='mkn')
p2, = plt.plot( size, mnk, color = color[1], linewidth = 3, label='mnk')
p3, = plt.plot( size, kmn, color = color[2], linewidth = 3, label='kmn')
p4, = plt.plot( size, knm, color = color[3], linewidth = 4, label='knm')
p5, = plt.plot( size, nmk, color = color[4], linewidth = 3, label='nmk')
p6, = plt.plot( size, nkm, color = color[5], linewidth = 3, label='nkm')
legend([p1, p2, p3, p4, p5, p6], ["mkn", "mnk", "kmn", "knm", "nmk", "nkm"])
lgd = legend(loc=9, bbox_to_anchor=(0.05,1))
plt.plot(loc='best', numpoints=1)

plt.grid(True)
savefig('Runtime.png')
```

## L1 D-cache Refs, L1 D-cache Misses, L2 D-cache Refs, L2 D-cache Misses, Hits 

```python
import numpy as np
import matplotlib.pyplot as plt

N = 6
L1D_hit_ratio = (46.4591, 3974.7322, 2981.2419, 1.8284, 18.3249, 126.4211)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, L1D_hit_ratio, width, color=color[1])

L2D_hit_ratio = (2.9597, 24.27, 17.516, 12.025, 12.3703, 33.7691)
rects2 = ax.bar(ind + width, L2D_hit_ratio, width, color=color[5])

# add some text for labels, title and axes ticks
ax.set_ylabel('Hit ratio')
ax.set_xlabel('Permutations')
ax.set_title('Hit ratio for the different permutations')
ax.set_xticks(ind + width)
ax.set_xticklabels(('mnk', 'mkn', 'kmn', 'nkm', 'knm', 'nmk'))

ax.legend((rects1[0], rects2[0]), ('L1 D-hit ratio', 'L2 D-hit ratio'))


def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

ax.set_ylim([0, 4500])
plt.grid(False)
savefig('hitratio.png')
```
