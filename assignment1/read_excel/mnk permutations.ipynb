```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline  
plt.rcParams['figure.figsize'] = (14.0, 8.0)
colours = []
colours.append( (26, 188, 156))
colours.append( (52, 152, 219))
colours.append( (155, 89, 182))
colours.append( (241, 196, 15))
colours.append( (231, 76, 60))
colours.append( (46, 204, 113))
colours.append( (230, 126, 34))
colours.append( (149, 165, 166))
colours.append( (52, 73, 94))
```

```python
mems = dict()
flops = dict()
with open('output.txt') as f:
    content = f.readlines()
    for c in content:
        if c[0] != '=':
            perm = c[-4:-1]
            values = dict()
            cc = [ ccc for ccc in c.split(' ') if ccc != '']
            print cc
            if perm not in mems:
                mems[perm] = []
            if perm not in flops:
                flops[perm] = []
                
            mems[perm].append(cc[0])
            flops[perm].append(cc[1])
```

```python
keys = mems.keys()
plt.title('flops/s vs memory occupation')
plt.xlabel('memory occupation (kB)')
plt.ylabel('flop/s')
k = 0
patches = []
for perm in keys:
    col = [ float(c) / 255.0 for c in colours[k]]
    plt.loglog( mems[perm], flops[perm], color = col, linewidth = 3)
    patch = mpatches.Patch(color = col, label=perm)
    patches.append( patch)
    
    k += 1
# plt.loglog( v_mem_lib, v_flops_lib, color = '#1abc9c', linewidth = 3)
plt.legend(handles=patches)
plt.grid(True)
```
